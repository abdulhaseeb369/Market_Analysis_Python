```python
#!pip install yfinance  #pip for installing packages

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import yfinance as yf #yahoo finance API
import seaborn as sea

print ("Imports completed")
```

```python
USDBondYield = yf.download("^TNX",period="5y",interval="1d")
USDBondYield.rename(columns = {"Adj Close":"USBondYeild"},inplace=True)

DowJones = yf.download("^DJI",period="5y",interval="1d")
DowJones.rename(columns = {"Adj Close":"DOW"},inplace=True)

Nasdaq = yf.download("^IXIC",period="5y",interval="1d")
Nasdaq.rename(columns = {"Adj Close":"NASDAQ"},inplace=True)

US500 = yf.download("^GSPC",period="5y",interval="1d")
US500.rename(columns = {"Adj Close":"US500"},inplace=True)

Dax = yf.download("^GDAXI",period="5y",interval="1d")
Dax.rename(columns = {"Adj Close":"DAX"},inplace=True)

Ftse = yf.download("^FTSE",period="5y",interval="1d")
Ftse.rename(columns = {"Adj Close":"FTSE"},inplace=True)

Cac = yf.download("^FCHI",period="5y",interval="1d")
Cac.rename(columns = {"Adj Close":"CAC"},inplace=True)

Nifty = yf.download("^NSEI",period="5y",interval="1d")
Nifty.rename(columns = {"Adj Close":"NIFTY"},inplace=True)

BankNifty = yf.download("^NSEBANK",period="5y",interval="1d")
BankNifty.rename(columns = {"Adj Close":"NIFTYBANK"},inplace=True)

Nikkei = yf.download("^N225",period="5y",interval="1d")
Nikkei.rename(columns = {"Adj Close":"NIKKEI"},inplace=True)

CrudeOil = yf.download("CL=F",period="5y",interval="1d")
CrudeOil.rename(columns = {"Adj Close":"CRUDE"},inplace=True)

Gold = yf.download("GC=F",period="5y",interval="1d")
Gold.rename(columns = {"Adj Close":"GOLD"},inplace=True)

USDINR = yf.download("INR=X",period="5y",interval="1d")
USDINR.rename(columns = {"Adj Close":"USDINR"},inplace=True)

```

```python
df = USDBondYield[['USBondYeild']]\
    .join(DowJones[['DOW']])\
    .join(Nasdaq[['NASDAQ']])\
    .join(US500[['US500']])\
    .join(Dax[['DAX']])\
    .join(Ftse[['FTSE']])\
    .join(Cac[['CAC']])\
    .join(Nifty[['NIFTY']])\
    .join(BankNifty[['NIFTYBANK']])\
    .join(Nikkei[['NIKKEI']])\
    .join(CrudeOil[['CRUDE']])\
    .join(Gold[['GOLD']])\
    .join(USDINR[['USDINR']])
    
```

```python
print (df)
```

```python
df.dropna(inplace=True)
```
